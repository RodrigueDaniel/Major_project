```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the CSV file
df = pd.read_csv(r'D:\Coding\Major-Project\new_\data\PEMfuel cell.csv')

# Display the first few rows to confirm the content
df.head()
```

```python
# Step 2: Convert all columns to numeric (non-numeric values will be set to NaN)
df_numeric = df.apply(pd.to_numeric, errors='coerce')

# Step 3: Drop rows with any NaN values
df_clean = df_numeric.dropna()

# Step 4: Compute the correlation matrix
correlation_matrix = df_clean.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .75})
plt.title('Correlation Matrix Heatmap for PEM Fuel Cell Data')
plt.tight_layout()
plt.show()
```
